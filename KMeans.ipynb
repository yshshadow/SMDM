```python
# train or load kmeans cluster model 
cluster = TwitterCluster()
```

```python
# the label of clusters that every tweets belongs to
labels = cluster.get_labels()
```

```python
df = pd.read_csv('./data/Tweets.csv')
df['label'] = labels.T
```
